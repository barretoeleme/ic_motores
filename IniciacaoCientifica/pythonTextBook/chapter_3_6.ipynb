```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    """Quickly make a DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    # data receives an iteration of values:
    # we first pick the first char (A) and combine it from
    # 0 to the number of indices (lines). we do that to
    # every cols letter
    return pd.DataFrame(data, ind)

# example DataFrame
make_df('ABC', range(3))
```

```python
# do NOT ask me what this does

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

# Simple Concatenation with pd.concat

```python
ser1 = pd.Series(['A', 'B', 'C'], index = [1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index = [4, 5, 6])
pd.concat([ser1, ser2])
```

```python
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])
display('df1', 'df2', 'pd.concat([df1, df2])')
```

```python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
display('df3', 'df4', "pd.concat([df3, df4], axis = 1)")
```

## Duplicate indices

```python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])
y.index = x.index # make duplicate indices
display('x', 'y', 'pd.concat([x, y])')
```

### Catching the repeats as an error

```python
# If you'd like to simply verify that the indices in
# the result of pd.concat() do not overlap, you can
# specify the verify_integrity flag.

try:
    pd.concat([x, y], verify_integrity = True)
except ValueError as e:
    print("Value error:", e)
```

### Ignoring the index

```python
display('x', 'y', 
        'pd.concat([x, y], ignore_index = True)')
# ingnores the datas indexes and forms its own
```

### Adding MultiIndex keys

```python
display('x', 'y', "pd.concat([x, y], keys = ['x', 'y'])")
```

## Concatenation with joins

```python
df5 = make_df('ABC', [1, 2])
df6 = make_df('BCD', [3, 4])
display('df5', 'df6', 'pd.concat([df5, df6])')
# creates a lot of NaN values
```

```python
display('df5', 'df6', 
        "pd.concat([df5, df6], join = 'inner')")
```
