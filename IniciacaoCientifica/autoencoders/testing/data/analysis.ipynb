```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_info(info, motor, target, measure, n = 30):

    n = 30
    if (measure == 'score'):
        index = info[f'{target}_{measure}'].nlargest(n).index
        which = "largest"
    else:
        index = info[f'{target}_{measure}'].nsmallest(n).index
        which = "smallest"

    infos = f"{target}_{measure}"

    figure = plt.scatter(info.layers[index], info[infos][index], 
                         c = info.neurons[index], s = info.learn_rate[index] * 3000, cmap = 'viridis', alpha = 0.9)

    figure = plt.colorbar().set_label('neurons')

    figure = plt.xlabel('layers')
    figure = plt.ylabel(infos)

    figure = plt.title(f'{target} loss prediction - Results & Parameters for {n} {which} {measure} - Motor {motor}')

    figure
```

```python
info_2D = pd.read_csv("motor_2D_info.csv")
info_Nabla = pd.read_csv("motor_Nabla_info.csv")
# info_V = pd.read_csv("motor_V_info.csv")
```

```python
plot_info(info_2D, '2D', 'hys', 'score')
```

```python
plot_info(info_2D, '2D', 'hys', 'mse')
```

```python
plot_info(info_2D, '2D', 'hys', 'mape')
```
