```python
def obs_vs_pred (true_value, predicted_value, title):
    plt.figure(figsize=(7, 7))
    plt.scatter(true_value, predicted_value, c='crimson')
    
    p1 = max(max(predicted_value), max(true_value))
    p2 = min(min(predicted_value), min(true_value))
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.xlim(-3, 5)
    plt.ylim(-3, 5)
    plt.xlabel('True Values', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.title(title, fontsize = 18)
    plt.show()
```

```python
obs_vs_pred(df_2D.loc[('cat_rand', 'hysteresis'), 'y_test'], df_2D.loc[('cat_rand', 'hysteresis'), 'y_pred'], "Motor 2D - cat_rand - hysteresis loss")
obs_vs_pred(df_2D.loc[('cat_rand', 'joule'), 'y_test'], df_2D.loc[('cat_rand', 'joule'), 'y_pred'], "Motor 2D - cat_rand - joule loss")

obs_vs_pred(df_2D.loc[('xgb_rand', 'hysteresis'), 'y_test'], df_2D.loc[('xgb_rand', 'hysteresis'), 'y_pred'], "Motor 2D - xgb_rand - hysteresis loss")
obs_vs_pred(df_2D.loc[('xgb_rand', 'joule'), 'y_test'], df_2D.loc[('xgb_rand', 'joule'), 'y_pred'], "Motor 2D - xgb_rand - joule loss")
```
