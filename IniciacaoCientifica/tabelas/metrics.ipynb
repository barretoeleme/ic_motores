```python
import pandas as pd
import numpy as np
```

```python
def read_result (motor, target, index, columns):
    THIS_PATH = f"../results/{motor}/{target}/results_"

    read = pd.DataFrame(index = index, 
                        columns = columns)

    read.loc['linear'] = pd.read_csv(f"{THIS_PATH}lin_reg.csv").iloc[0]
    read.loc['reg_tree'] = pd.read_csv(f"{THIS_PATH}reg_tree.csv").iloc[0]
    read.loc['rand_for'] = pd.read_csv(f"{THIS_PATH}rand_for.csv").iloc[0]
    read.loc['catboost'] = pd.read_csv(f"{THIS_PATH}catboost.csv").iloc[0]
    read.loc['xgboost'] = pd.read_csv(f"{THIS_PATH}xgboost.csv").iloc[0]
    read.loc['cat_rand'] = pd.read_csv(f"{THIS_PATH}cat_rand.csv").iloc[0]
    read.loc['xgb_rand'] = pd.read_csv(f"{THIS_PATH}xgb_rand.csv").iloc[0]

    return read
```

```python
methods = ['linear', 
           'reg_tree', 
           'rand_for', 
           'catboost', 
           'xgboost',
           'cat_rand',
           'xgb_rand']
metrics = ['score', 
           'mse',
           'mape']
targets = ['hysteresis', 'joule']
```

```python
index = pd.MultiIndex.from_product([methods, targets], names = ['method', 'target'])
columns = pd.Index(metrics, name = 'metric')

df_2D = pd.DataFrame(index = index, columns = columns)
```

```python
read_2D = pd.DataFrame(index = index, columns = columns)

for method in methods:
    for target in targets:
        results = read_result("2D", target, methods, columns)
        read_2D.loc[(method, target)] = results.loc[method]

read_2D
```

```python
read_Nabla = pd.DataFrame(index = index, columns = columns)

for method in methods:
    for target in targets:
        results = read_result("Nabla", target, methods, columns)
        read_Nabla.loc[(method, target)] = results.loc[method]

read_Nabla
```

```python
read_V = pd.DataFrame(index = index, columns = columns)

for method in methods:
    for target in targets:
        results = read_result("V", target, methods, columns)
        read_V.loc[(method, target)] = results.loc[method]

read_V
```
